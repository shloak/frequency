# Kay Experiments

```python
SNRdB = 2
exps = [8, 10, 12, 14, 16]
ours = np.load('../data/kay_testing/our_method_params_snr{}.npy'.format(SNRdB))
succ = np.load('../data/kay_testing/successive_est_params_snr{}.npy'.format(SNRdB))
for i in range(len(exps)):
    nums = [i for i in range(2, exps[i] + 1)]
    
    fig = plt.figure(figsize=(10,4))
    ax = fig.add_subplot(111)
    ax.set_xlabel('Total Samples Taken') 
    ax.set_ylabel('Accuracy')
    ax.plot([exps[i] * n for n in nums], succ[i], '-bo', label='Successive Estimation Kay')
    ax.plot([exps[i] * 2 * n for n in nums], ours[i], '-ro', label='Our method')
    ax.set_title('Accuracy vs total samples taken at snr {}dB and 2 ^ {} signal length'.format(SNRdB, exps[i]))
    ax.legend(loc='best',
               ncol=3)
    plt.show()
```
