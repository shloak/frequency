```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math
```

Results
 - MLE frequency detection works much better with random m samples compared to first m
 - Divide and conquer method using neural network does not scale well
   - Works when num possible frequencies is in magnitude of 100s & receiving >logarithmic samples at each subsampling rate
       - Works at sub 0 SNR
   - When num possible frequencies is greater than 100s, binary detector becomes bad -> can't detect frequency 
       - Had error in code last time which made it seem good

```python
# mle frequency accuracy at a given snr
# first m samples, random m samples, m subsampled samples set
# N=27000 possible freq, m=300 samples

plt.plot(np.load('../data/freq_detect/snrs.npy'), np.load('../data/freq_detect/mle_first.npy'), '--bo', label='mle first 300 samples')
plt.plot(np.load('../data/freq_detect/snrs.npy'), np.load('../data/freq_detect/mle_random.npy'), '--ro', label='mle random 300 samples')
plt.plot(np.load('../data/freq_detect/snrs.npy'), np.load('../data/freq_detect/mle_2power.npy'), '--go', label='mle 300 samples at 2^n')
plt.xlabel('snr db', fontsize=18)
plt.ylabel('testing accuracy', fontsize=18)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.show()
```

```python
# divide conquer frequency accuracy at a given snr
# changed m = samples at each subsampling phase
# N=1024 possible freq, snr=-2 db

plt.plot(np.load('../data/divide_conquer/ms.npy'), np.load('../data/divide_conquer/bin_accs.npy'), '--bo', label='top/bottom detection')
plt.plot(np.load('../data/divide_conquer/ms.npy'), np.load('../data/divide_conquer/freq_accs.npy'), '--ro', label='frequency detection')
plt.xlabel('m=num samples', fontsize=18)
plt.ylabel('testing accuracy', fontsize=18)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.show()
```
