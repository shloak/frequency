```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math
```

Results

```python
# divide conquer frequency accuracy at a given snr
# changed m = samples at each subsampling phase
# N=1024 possible freq, m=10 (logn) samples at each subsampling (starting at 0), 6 layers

plt.plot(np.load('../data/divide_conquer/snrs_1024shift.npy'), np.load('../data/divide_conquer/snr_acc_binary_1024shift.npy'), '--bo', label='top/bottom detection')
plt.plot(np.load('../data/divide_conquer/snrs_1024shift.npy'), np.load('../data/divide_conquer/snr_acc_frequency_1024shift.npy'), '--ro', label='frequency detection')
plt.xlabel('snr db', fontsize=18)
plt.ylabel('testing accuracy', fontsize=18)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.show()
```
