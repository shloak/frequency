```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math
```

```python
# 10 trials at each snr
# 3 conv layers [[5, 2], [3, 2], [3, 2]] to one softmax layer
# kay value calculated by getting threshhold that would give max accuracy
# N=27000 possible freq, m=300 samples

plt.plot(np.load('./data/singleton/snrs.npy'), np.load('./data/singleton/nn_accs.npy'), '--bo')
plt.plot(np.load('./data/singleton/snrs.npy'), np.load('./data/singleton/kay_accs.npy'), '--ro')
plt.xlabel('snr db', fontsize=18)
plt.ylabel('testing accuracy', fontsize=18)
plt.show()
```

```python
# 10 trials at each layer count
# adding layers of size [3, 2] after first [5, 2] layer to one softmax layer
# N=27000 possible freq, m=300 samples

plt.plot(np.load('./data/singleton/layers.npy'), np.load('./data/singleton/layer_accs.npy'), '--bo')
plt.xlabel('# conv layers', fontsize=18)
plt.ylabel('testing accuracy', fontsize=18)
plt.show()
```

```python
# 10 trials at each layer count
# using training set of 500 signals of set snr
# N=27000 possible freq, m=300 samples

plt.plot(np.load('./data/singleton/snrs_training.npy'), np.load('./data/singleton/nn_accs_snrs_training.npy'), '--bo')
plt.xlabel('snr db of training set', fontsize=18)
plt.ylabel('testing accuracy (snr = -6)', fontsize=18)
plt.show()
```
